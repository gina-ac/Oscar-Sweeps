```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
# parse data from html, to a beautifulsoup object, to a dataframe

# create empty dataframe
df_complete = pd.DataFrame()

#loop through all awards ceremonies
for x in range(1, 96):
    # format url
    y = 'th'
    if str(x)[-1] == '1' and x != 11:
        y = 'st'
    elif str(x)[-1] == '2' and x != 12:
        y = 'nd'
    elif str(x)[-1] == '3' and x != 13:
        y = 'rd'    
    url = f"https://en.wikipedia.org/wiki/{x}{y}_Academy_Awards"
    try:
        # get the response in the form of html
        response = requests.get(url)
        soup = BeautifulSoup(response.text, 'html.parser')
        element = 'table'
        attribute = 'class'
        attribute_name = 'wikitable'
        tables = soup.find_all(element, {attribute:attribute_name})
        multiple_wins = tables[2]
        data = pd.read_html(str(multiple_wins))
        # convert list to dataframe
        df = pd.DataFrame(data[0])
        df['Ceremony'] = x
        # concatenate dataframes
        df_complete = pd.concat([df_complete, df], ignore_index=True)
        print(f"Complete: {url}")
    except:
         print(f"Error: {url}")
```

```python
# add Year column
df_multiple_wins_temp4['Year'] = df_multiple_wins_temp4['Ceremony'] + 1928   # The first ceremony was in 1929.
```

```python
# add Decade column
bins = [1920, 1929, 1939, 1949, 1959, 1969, 1979, 1989, 1999, 2009, 2019, 2029]
labels = ['1920', '1930', '1940', '1950', '1960', '1970', '1980', '1990', '2000', '2010', '2020']
df_multiple_wins_temp4['Decade'] = pd.cut(x=df_multiple_wins_temp4['Year'], bins=bins, labels=labels)
```

```python
# create, sort, fix data type, and re-index final dataframe
df_wins = df_multiple_wins_temp4
df_wins = df_wins.astype({'Wins': int})
df_wins = df_wins.sort_values(by=['Ceremony', 'Wins', 'Film'], ascending=[True, False, True])
df_wins = df_wins.reset_index(drop=True)

# turn off truncating
pd.set_option('display.max_rows', None)
```

```python
# SPLIT INTO BEFORE AND AFTER CATEGORY MAXIMUM EXPANSIONS
# PULL FROM BEST ACTOR and BEST SUPPORTING FOR TONIGHT SHOW APPEARANCES AND HOW CLOSE TO OSCARS THEY APPEARED
```

```python
# plot total multiple award winning films by decade
import matplotlib.pyplot as plt

x = df_wins['Decade'].unique()
y = df_wins['Decade'].value_counts().sort_index()
plt.plot(x, y)
for i in range(len(x)): 
    plt.annotate(y[i], (x[i], y[i] - 2)) 
plt.show()
```

```python
# plot multiple award winning films by year
x = df_wins['Year'].unique()
y = df_wins['Year'].value_counts().sort_index()

plt.plot(x, y)
plt.show()
```

```python
# plot total wins by multi-winners by decade
x = df_wins['Decade'].unique()
y = df_wins.groupby('Decade')["Wins"].sum()
plt.plot(x, y)
for i in range(len(x)): 
    plt.annotate(y[i], (x[i], y[i] - .1)) 
plt.show()
```

```python
# plot average wins per decade
x = df_wins['Decade'].unique()
y = df_wins.groupby('Decade')["Wins"].mean().round(1)
plt.plot(x, y)
for i in range(len(x)): 
    plt.annotate(y[i], (x[i], y[i] - .1)) 
plt.show()
```

```python
# plot average multi-winning films per year by decade
x = df_wins['Decade'].unique()
y1 = df_wins.groupby('Decade')['Film'].nunique()
y2 = df_wins.groupby('Decade')['Year'].nunique()
y = (y1 / y2).round(1)
# y = df_wins.groupby('Decade')["Films"].mean().round(1)
plt.plot(x, y)
for i in range(len(x)): 
    plt.annotate(y[i], (x[i], y[i] - .1)) 
plt.show()
```

```python
# plot average wins per year
x = df_wins['Year'].unique()
y = df_wins.groupby('Year')['Wins'].mean()
plt.bar(x, y)
plt.show()
```

```python
# plot average wins per year by decade
x = df_wins['Decade'].unique()
y1 = df_wins.groupby('Decade')['Wins'].sum()
y2 = df_wins.groupby('Decade')['Year'].nunique()
y = (y1 / y2).round(1)
plt.plot(x, y)
for i in range(len(x)): 
    plt.annotate(y[i], (x[i], y[i] - .1)) 
plt.show()
```
